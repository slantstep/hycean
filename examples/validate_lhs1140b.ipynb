# LHS 1140 b Pipeline Validation

Compare pipeline results to:
1. Your manual analysis from yesterday
2. Cadieux+2024 published results
3. Damiano+2024 published results

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load pipeline results
with open('../data/outputs/lhs1140b_pipeline_test.json', 'r') as f:
    pipeline_results = json.load(f)

# Load config with published results
with open('../configs/planets/lhs1140b.json', 'r') as f:
    config = json.load(f)

published = config['published_results']
```

```python
# Comparison table
print("Pipeline Validation Results")
print("="*60)
print(f"{'Metric':<40s} {'Pipeline':<15s} {'Published'}")
print("="*60)

# Atmosphere detection
print(f"{'Atmosphere detected':<40s} {str(pipeline_results['spectroscopy']['atmosphere_detected']):<15s} {published['atmosphere_detected']}")

# Rayleigh scattering
rayleigh_pipe = pipeline_results['spectroscopy']['rayleigh']['consistent_with_scattering']
rayleigh_pub = published['rayleigh_scattering']['detected']
print(f"{'Rayleigh scattering':<40s} {str(rayleigh_pipe):<15s} {rayleigh_pub}")

# Rayleigh significance
alpha = pipeline_results['spectroscopy']['rayleigh']['alpha']
alpha_err = pipeline_results['spectroscopy']['rayleigh']['alpha_error']
if alpha and alpha_err:
    our_sigma = abs(alpha - (-4)) / alpha_err
    pub_sigma = published['rayleigh_scattering']['significance']
    print(f"{'Rayleigh detection significance':<40s} {our_sigma:.1f}σ{' '*9} {pub_sigma}σ")

# H2 ruled out
h2_ruled = published['molecules']['H2']['ruled_out_sigma']
print(f"{'H2 envelope ruled out':<40s} {'N/A (no fit)':<15s} >{h2_ruled}σ")

print("="*60)
```

## Validation Criteria

✓ **PASS** if:
- Atmosphere detected by both
- Rayleigh scattering detected by both
- Our significance within 2σ of published
- Molecular features not detected by either (consistent with low S/N)

⚠ **REVIEW** if:
- Quantitative differences but qualitative agreement

✗ **FAIL** if:
- Opposite conclusions (e.g., we detect atmosphere, they don't)
- Major discrepancies (>5σ difference in slope)
